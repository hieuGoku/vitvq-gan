```python
from glob import glob
from tqdm import tqdm
folders = glob("fastMRI_brain_DICOM/*/", recursive = True)
folders
```

```python
files = [glob(f"{folders[i]}/*.*", recursive = True) for i in tqdm(range(len(folders)))]
files
```

```python
import pydicom
import matplotlib.pyplot as plt
def convert_dicom_to_png(dicom_file, png_file):
    # Load the DICOM file
    ds = pydicom.dcmread(dicom_file)

    # Extract the pixel data from the DICOM dataset
    pixel_data = ds.pixel_array

    # Resize the pixel data if needed
    # You can use interpolation methods like plt.imshow(pixel_data, cmap='gray', interpolation='bilinear')

    # Save the pixel data as a PNG file
    plt.imsave(png_file, pixel_data, cmap='gray')
```

```python
import os
png_folder = "fastMRI_brain_PNG"
if not os.path.exists(png_folder):
    os.makedirs(png_folder)
```

```python
for i in tqdm(range(len(files))):
    os.makedirs(folders[i].replace("DICOM","PNG"))
    for j in range(len(files[i])):
        convert_dicom_to_png(files[i][j], f"{files[i][j].replace('DICOM','PNG').replace('dcm', 'png')}")
```
